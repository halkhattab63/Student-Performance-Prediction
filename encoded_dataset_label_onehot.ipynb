```python
# Gerekli kütüphaneleri import etimport pandas as pd

from sklearn.preprocessing import LabelEncoder
from google.colab import files
```

```python
# CSV dosyasını bilgisayarından yükle

uploaded = files.upload()  # Dosya seçmek için bir pencere açılacak
file_name = list(uploaded.keys())[0]
```

```python
# Veriyi oku

df = pd.read_csv(file_name)
```

```python
# Pandas ayarlarını yaparak tüm satırları ve sütunları görüntüle

pd.set_option('display.max_rows', None)     # Tüm satırları görüntüle
pd.set_option('display.max_columns', None)  # Tüm sütunları görüntüle
pd.set_option('display.width', None)        # Genişlik sınırlamasını kaldır
pd.set_option('display.max_colwidth', None) # Uzun metinleri tamamen göster
```

```python
# Orijinal veriyi görüntüle (isteğe bağlı)

print("Orijinal Veri:")
print(df.to_string())
```

```python
# Kategorik sütunları belirle

categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
# Label Encoding işlemi yap

df_label_encoded = df.copy()
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df_label_encoded[col] = le.fit_transform(df_label_encoded[col])
    label_encoders[col] = le
```

```python
# Label Encoding sonrasında tüm veriyi görüntüle

print("\n Label Encoded Veri (Tüm Satırlar ve Sütunlar):")
print(df_label_encoded.to_string())
```

```python
# One-Hot Encoding işlemi yap

df_one_hot_encoded = pd.get_dummies(df.copy(), columns=categorical_columns)
```

```python
# One-Hot Encoding sonrasında tüm veriyi görüntüle

print("\n One-Hot Encoded Veri (Tüm Satırlar ve Sütunlar):")
print(df_one_hot_encoded.to_string())
```

```python
# Düzenlenmiş verileri CSV dosyasına kaydet

df_label_encoded.to_excel("label_encoded_dataset.xlsx", index=False)
df_one_hot_encoded.to_excel("one_hot_encoded_dataset.xlsx", index=False)
```

```python
# Düzenlenmiş dosyaları bilgisayarına indir

files.download("label_encoded_dataset.xlsx")
files.download("one_hot_encoded_dataset.xlsx")
```
